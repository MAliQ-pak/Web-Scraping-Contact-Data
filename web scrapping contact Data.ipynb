```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.canada411.ca/search/?stype=si&what=&where=Ontario'

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all(class_ ="listing__row")
```

```python
for i in element:
    print(i.text.splitlines()[1:])
```

```python
rows = []

for i in element:
    lines = i.text.splitlines()[1:]
    filtered_lines = [line for line in lines if line.strip()]
    rows.append(filtered_lines)
    
rows
```

```python
contact_numbers = []

for i in range(0, len(rows), 2):
    name_num = rows[i]
    contact_numbers.append(name_num)
    
contact_numbers
```

```python
df = pd.DataFrame(contact_numbers, columns=['Name', 'Contact Number'])
df
```

```python
addresses = []

for i in range(1, len(rows), 2):
    address = rows[i][0]
    addresses.append(address)

addresses
```

```python
df['Address'] = addresses
```
